```python
import matplotlib.pyplot as plt
import numpy as np
```

# Funciones Trascendentes

Todas aquellas funciones que no pueden ser expresadas con polinomios

### Funciones Trigonometricas

```python
N= 100

x=np.linspace(-10,10,num=N)

def f(x):
    return np.cos(x)

y = f(x)

plt.plot(x,y)

```

```python
def f(x):
    return np.sin(x)

y = f(x)

plt.plot(x,y)
```

```python
def f(x):
    return np.tan(x)

y = f(x)

plt.plot(x,y)
```

# Funciones Exponenciales

```python
def f(x):
    return np.exp(x)

y = f(x)

plt.plot(x,y)
```

# Funcion Logaritmo

```python
def f(x):
    return np.log2(x) 

x = np.linspace(0.001,256,num=1000)
plt.plot(x,f(x))
```
